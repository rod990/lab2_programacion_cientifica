```python
from functools import reduce
import random
#actividad_1
num=[2,4,6]
ecuacion= list(map(lambda x:x**2+2*x+1,num))
print(ecuacion)
#actividad_2
velocidades = [10, 20, 30, 40, 50]
convertir=list(map(lambda x:x/3.6,velocidades))
print(convertir)
#actividad_3
valores = [2,5,8,11,14,17,20]
par=list(filter(lambda x:x%2==0,valores))
print(par)
#actividad_4
numeros = [2,3,4,5]
resultado = reduce(lambda x,y: x+y, numeros)
print (resultado)
#actividad_5
data = [12,15,22,35,41,8,3,55,62,90]
resultado_data= reduce(lambda x,y: x+y,map(lambda x:x*2,filter(lambda x: x>10,data)))
print(resultado_data)
#actividad_6
data_grande = [random.randint(0,100) for _ in range(1000000)]
procesado=list(map(lambda x: x**2,filter(lambda x:x>50,data_grande)))
promedio= sum(procesado)/len(procesado)
print(promedio)
#desafio
sensores = [random.uniform(10,80) for _ in range(500000)]
n=list(map(lambda x:x+273.15,filter(lambda x:10<=x<=70,sensores)))
promedio_n= sum(n)/len(n)
print(promedio_n)
```
